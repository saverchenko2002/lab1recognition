```python
import os
import librosa
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import models, layers, optimizers
import matplotlib.pyplot as plt
import librosa.display
from sklearn.preprocessing import LabelEncoder
```

```python
import librosa.display
max_length=130
# Загрузка аудиофайла и его расшифровки, включая создание спектрограммы
def load_data(audio_path, verbose):
    audio_files = []
    transcripts = []

    for audio_file in os.listdir(audio_path):
        if(verbose):
            print(audio_file)
        audio, sr = librosa.load(os.path.join(audio_path, audio_file), sr=None)
        # print(sr)
        # print(audio.shape)
        spectrogram = librosa.feature.melspectrogram(y=audio)
        # print(spectrogram.shape)
        spectrogram = librosa.power_to_db(spectrogram, ref=np.max)  # Преобразование в децибелы
        # print(spectrogram.shape)

        audio_files.append(spectrogram)
        transcripts.append(audio_file)
        
    # max_length = max(spec.shape[1] for spec in audio_files)
    
    audio_files_padded = []
    
    for i in range(len(audio_files)):
        audio_empty = np.zeros((128, max_length))
        audio_empty[:, :audio_files[i].shape[1]] = audio_files[i]
        audio_files_padded.append(audio_empty)
        
    
    



    # Построение плота
    plt.figure(figsize=(15, 10))

    for i in range(min(4, len(audio_files_padded))):  # Ограничиваем количество спектрограмм до 4
        plt.subplot(2, 2, i+1)
        librosa.display.specshow(audio_files_padded[i], sr=sr, x_axis='time', y_axis='mel')
        plt.colorbar(format='%+2.0f dB')
        plt.title('Spectrogram {}'.format(os.listdir(audio_path)[i]))

    plt.tight_layout()
    plt.savefig('spectrograms.png')
    plt.show()
    
    
    
    
    # audio_files_padded = pad_sequences(audio_files, padding='post', dtype='float32', maxlen=max_length)
    # print(audio_files_padded)

    return np.array(audio_files_padded), np.array(transcripts)
```

```python
def create_and_train_model(X_train, y_train):
    # Здесь создайте вашу нейронную сеть, используя, например, TensorFlow или PyTorch
    # Пример TensorFlow:
    model = models.Sequential([
        layers.Flatten(input_shape=(X_train.shape[1], X_train.shape[2])),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')  # Здесь уточните количество классов
    ])
    model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
                  loss='binary_crossentropy',  # Здесь уточните тип потерь
                  metrics=['accuracy'])

    model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

    return model
```

```python
audio_path = "initial phrases voiced"
initial_phrases_txt_path = "initial_phrases.txt"

initial_phrases = []

with open(initial_phrases_txt_path, 'r', encoding='utf-8') as file:
    initial_phrases = file.read().split('\n')
```

```python
X_train, y_train = load_data(audio_path, False);
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Создание модели CTC
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.1))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(len(np.unique(y_train_encoded)) + 1, activation='softmax'))  # +1 для учета пустого символа в CTC

# Компиляция модели
from tensorflow.keras import optimizers
model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

X_train = X_train / 255.0

# Обучение модели
model.fit(X_train, y_train_encoded, epochs=20, batch_size=1, verbose=1)
```

```python
X_test, _ = load_data('cross phrases voiced', True)
```

```python
predictions = model.predict(X_test/255.0)
predicted_labels = np.argmax(predictions, axis=1)
predicted_labels_text = label_encoder.inverse_transform(predicted_labels)
print(predicted_labels_text)
```
