```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np  # добавлен импорт NumPy

# Загрузка аудиофайла
file_path = 'initial phrases voiced/Быстрая лошадь.mp3'
y, sr = librosa.load(file_path, sr=None)

# Получение спектрограммы
spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# Визуализация спектрограммы
librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max), y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()

spectrogram.shape
```

```python
import librosa
import librosa.display
import numpy as np
from PIL import Image

# Загрузка аудиофайла
file_path = 'initial phrases voiced/Быстрая лошадь.mp3'
y, sr = librosa.load(file_path, sr=None)

# Получение спектрограммы
spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)

# Преобразование в децибелы и нормализация
spectrogram_db = librosa.power_to_db(spectrogram, ref=np.max)
normalized_spectrogram = (spectrogram_db - np.min(spectrogram_db)) / (np.max(spectrogram_db) - np.min(spectrogram_db))

# Преобразование в изображение
image_array = (normalized_spectrogram * 255).astype(np.uint8)
print(image_array.shape)
image = Image.fromarray(image_array)

# Сохранение изображения или использование в обучении нейронной сети
image.save('спектрограмма.png')
```
